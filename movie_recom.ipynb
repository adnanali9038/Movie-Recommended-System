```python
import numpy as np
import pandas as pd
df=pd.read_csv("ml-100k/u.data",sep='\t')
df.head()
```

```python
column_names=["userid","itemid","rating","timestamp"]
df=pd.read_csv("ml-100k/u.data",sep='\t',names=column_names)
df.head()
```

```python
df['userid'].nunique()
df['itemid'].nunique()
```

```python
movie_title=pd.read_csv("ml-100k/u.item",sep="\|",header=None)
movie_title.shape
movie_title=movie_title[[0,1]]
movie_title.columns=['itemid','title']
movie_title.head()
```

```python
df=pd.merge(df,movie_title,on="itemid")
df.head()
```

```python
import matplotlib.pyplot as plt
df.groupby('title').mean()['rating']
df.groupby('title').mean()['rating'].sort_values(ascending=False)
df.groupby('title').count()['rating'].sort_values(ascending=False)
```

```python
ratings=pd.DataFrame(df.groupby('title').mean()['rating'])
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title').count()['rating'])
ratings.sort_values(by='rating',ascending=False)
```

```python
moviemat=df.pivot_table(index="userid",columns="title",values="rating")
moviemat.head()
```

```python
related_ratings=moviemat['187 (1997)']
similar_to_187=moviemat.corrwith(related_ratings)
```

```python
corr_187=pd.DataFrame(similar_to_187,columns=['correlation'])
corr_187.dropna(inplace=True)
corr_187.head()
```

```python
corr_187.sort_values('correlation',ascending=False).head(10)
```

```python
corr_187=corr_187.join(ratings['num of ratings'])
corr_187.head()
```

```python
corr_187[corr_187['num of ratings']>100].sort_values('correlation',ascending=False)
```

```python
def predict_movies(moviename):
    movie_user_rating=moviemat[moviename]
    similar_to_movie=moviemat.corrwith(movie_user_rating)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie=corr_movie.join(ratings['num of ratings'])
    predictions=corr_movie[corr_movie['num of ratings']>10].sort_values('correlation',ascending=False)
    return predictions
```

```python
predictions=predict_movies("Titanic (1997)")
predictions.head()
```
